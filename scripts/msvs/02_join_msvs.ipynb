```python
from os.path import exists
import numpy as np
import pandas as pd
from information.terms_info import tags
```

```python
df = pd.DataFrame()

for tag in tags:
    normalize_tag = tag.replace('/', '.')
    filename = f"./../../data/processed/msvs/{normalize_tag}.csv"

    if not exists(filename):
        continue

    new_df = pd.read_csv(filename, parse_dates=['date'])
    new_df.rename(columns={'index': tag}, inplace=True)

    print('[i]', f"{tag=}", f"{df.shape=}", f"{new_df.shape=}")

    if len(df) == 0:
        df = new_df.copy()
        continue
    else:
        df = pd.merge(
            df, new_df,
            how="outer",
            left_on="date", right_on="date",
            suffixes=("", ""))

df.fillna(0, inplace=True)

df.to_pickle('./../../data/processed/gt.pkl')
df.to_csv('./../../data/processed/gt.csv', index=False)
```
